##Creating Arrays in Numpy

```python
import numpy as np

arr_1D = np.array([1,2,3,4,5])
print("1D array:", arr_1D)

arr_2D = np.array([[1,2,3,4,4],[5,6,7,8,9]])
print("2D Array:", arr_2D)
```

```python
#Comparison between  numpy array and list  why numpy array is fasrer then  numpy list 

import time
start = time.time()
np_list = [i*2 for i in range(100000)]
print("\n List operation time: " , time.time() - start )

# NUMPY ARRAY
start = time.time()

np_array = np.arange(0 ,100000)*2
print("\n Array operation time: " , time.time() - start )

```

CREATING ARRAS OF ALL TYPE IN NUMPY


```python
np_Zeros = np.zeros((3,4))
print("Zeros Array\n" , np_Zeros)

np_ones = np.ones((2,3)) 
print("Ones Array\n" , np_ones)

full = np.full((2,2),7)
print("Full Array\n" , full)

random = np.random.random((2,3))
print("Random Array\n" , random)

sequence = np.arange(0,12,2)
print("Sequence Array\n" , sequence)

```

VECTOR  MATRIX AND TENSOR 


```python
Vector = np.array([1,2,3,4])
print("Vector" , Vector)

Matrix = np.array([[1,2,3,4],
                  [2,4,5,6]])
print("Matrix" , Matrix)

tensor = np.array([
    [
        [[1,2,3],[2,6,7]],
        [[1,2,4],[7,8,9]]
    ]
])
print("Tensor", tensor)
```
